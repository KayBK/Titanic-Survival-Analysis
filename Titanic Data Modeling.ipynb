```python
#Handling Missing Values Age,Cabin and Embarked

```

```python
df["Age"].fillna(df["Age"].median(), inplace=True)
```

```python
df['Embarked'].value_counts()
```

```python
#Since only 2 records are missing in the Embarked column,fill them with the most frequent embarkation port.
df["Embarked"].fillna(df["Embarked"].mode()[0], inplace=True)
```

```python
#Create a Missing Indicator for Cabin
df["Cabin_known"] = df["Cabin"].notna().astype(int)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df["Embarked"] = df["Embarked"].replace({
    "S": "Southampton",
    "C": "Cherbourg",
    "Q": "Queenstown"
})
```

```python
df.head()
```

```python
#create a column age group
bins = [0, 12, 19, 35, 55, 80]
labels=['Child','Teenager','Young-Adult','Adult','Senior']
df['Age_group']=pd.cut(df['Age'],bins=bins,labels=labels,include_lowest=True)
```
